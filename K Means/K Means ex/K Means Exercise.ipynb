# K Means
In this notebbok we apply the K Means clustering algorithm to an artificial dataset created with sklearn

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Data

```python
from sklearn.datasets import make_blobs
```

Generate artificial data for clustering using sklearn. Using the make_blobs function, it is possile to decide the number of samples, features and number of centers of the dataset

```python
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 101)
```

Plot the data

```python
sns.set_style('whitegrid')
plt.rcParams["patch.force_edgecolor"] = True
plt.scatter(data[0][:,0],data[0][:,1], c = data[1], cmap = 'rainbow')
```

## K Means Algorithm

```python
from sklearn.cluster import KMeans
```

```python
# Create an instance of KMeans
kmeans = KMeans(n_clusters = 4)
kmeans.fit(data[0])
```

```python
# Cluster centers
kmeans.cluster_centers_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize = (10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_, cmap = 'rainbow')

ax2.set_title('Originals')
ax2.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'rainbow')
```

If we look at the 2 plots, without paying attention to the 2 colors that are inverted we can notice the differenc, between the original clusters and the clusters after the k means algorithm
